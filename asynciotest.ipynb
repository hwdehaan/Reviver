```python
import asyncio
import ipywidgets as widgets
import random
import string
import numpy as np
#from IPython.display import clear_output

wtop = widgets.Output(layout={'border': '1px solid black', 'height': '50px'})
#wout = widgets.Output(layout={'border': '1px solid black', 'height': '400px'})
wout = widgets.HTML(layout={'border': '1px solid black', 'height': '400px'})
display(wtop,wout)

loop = asyncio.get_event_loop()

ISgoevent = asyncio.Event()



#rng = np.random.randint(0,50)
#li_stest = list(stest)
#li_stest[3] = string.ascii_letters[rng]
#stest=''.join(li_stest)

async def ftest():
    global ISgoevent
    global stest
    spre = '<div style="font-family: \'Courier\', sans-serif; font-size: 16px; color: black; font-weight: normal;">'
    spos = '</div>'
    while True:
        i=0
        stest=''
        for ilett in range(0,50):
            stest += random.choice(string.ascii_letters)    
        while i < 20:
            rng = np.random.randint(0,50)
            li_stest = list(stest)
            li_stest[3] = string.ascii_letters[rng]
            stest=''.join(li_stest)            
            if ISgoevent.is_set():
                #wout.value = f"<pref>{i}</pre>"
                wout.value = f"<pref>{spre} {stest} {spos}</pre>"
                await asyncio.sleep(0.15)
                i += 1
            else:
                while not ISgoevent.is_set():
                    #wout.value = f"<pref>dur  {i}</pre>"
                    await asyncio.sleep(0.25)
                    #print(i)

global ttest 
ttest = None

def runbutclick(button_instance):
    global ttest
    ISgoevent.set()
    if ttest == None or ttest.done():
        ttest = loop.create_task(ftest())
    else:
        print('already running')
def stopbutclick(button_instance):
    global ttest
    global ISgoevent
    ttest.cancel()
def pausebutclick(button_instance):
    global ttest
    global ISgoevent
    if ISgoevent.is_set():
        ISgoevent.clear()
    else:
        print('unpause')
        ISgoevent.set()
        
##########################################################
run_button = widgets.Button(
    description = 'run'
)
run_button.on_click(runbutclick) 
stop_button = widgets.Button(
      description = 'stop'
)
stop_button.on_click(stopbutclick)
pause_button = widgets.Button(
      description = 'pause'
)
pause_button.on_click(pausebutclick)
##########################################################

def main_loop():
    hbox = widgets.HBox([run_button, stop_button,pause_button])
    wtop.append_display_data(hbox) 

if __name__ == "__main__":   
    main_loop()

```

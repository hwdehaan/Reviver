# Process Mask

```python

def fmask(img_name):
    #img_name = 'testcrazy2.bmp'
    
    img = IPIL.open(img_name) # Load image
    
    imgarray = asarray(img)
    
    print(type(imgarray))
    print(imgarray.shape) 
    
    imgrmax=int(imgarray.shape[0])
    imgcmax=int(imgarray.shape[1])

    rmax=int(imgarray.shape[0]/cgsizerow)
    cmax=int(imgarray.shape[1]/cgsizecol)


    print(str(imgrmax) + " " + str(imgcmax))
    print(str(rmax) + " " + str(cmax))

    mask=np.zeros((rmax+1,cmax+1),dtype=int)
    
    rowmask = 0
    colmask = 0
    
    for row in range(0,imgrmax):
        if row % cgsizerow == 0:
            colmask = 0
            for col in range(0,imgcmax):            
                if col%cgsizecol==0:               
                    if imgarray[row,col,0] > 128:
                        #print(str(rowmask) + " " + str(colmask))
                        mask[rowmask,colmask]=0
                        pix = 0
                    else:
                        pix = 1
                        mask[rowmask,colmask]=1
    
                    colmask += 1
                    #print(pix,end='')
            rowmask+=1
            #print('')

    return mask, rmax, cmax
    
    #rmax=mask.shape[0] - 1
    #cmax=mask.shape[1] - 1
```

# Make text and write HTML

```python
def fmake(fout,poem):
    
    
    fout.write('<HTML>\n')
    
    fout.write('<HEAD>\n')
    fout.write('<style> h1 {font-size: ' + str(nfontsize) + 'em; font-family: courier; font-style: regular; font-weight: normal;)</style>\n')
    fout.write('</HEAD>\n')
    
    fout.write('<BODY>\n')
    fout.write('<H1>\n')

    
    curr_shape=0
    
    full_text=[]

    mswords = poem.split()
    Nwords = len(mswords)
    i_inword = np.zeros(Nwords,dtype=int)
    
    
    iword = 0
    for kk in range(0,Nbigrows):
        for i in range(0,rmax):
            this_line=''
            for k in range(0,Nbigcols):           
                for j in range(0,len_ext_fill):
                    this_line = this_line + fill_ext
                for j in range(0,cmax):
                    itile = kk*Nbigcols+k
                    while itile >= Nwords:
                        itile -= Nwords
                    if mask[i,j] == 0:
                        this_line = this_line + fill_int
                    else:
                        this_line = this_line + mswords[itile][i_inword[itile]]
                        i_inword[itile] += 1
                        if i_inword[itile] >= len(mswords[itile]):
                            i_inword[itile]=0
                for j in range(0,len_ext_fill):
                    this_line = this_line + fill_ext
            print(this_line)
            full_text.append(this_line)
    
            
            for ijk in range(0,len(this_line)):
                fout.write(this_line[ijk])
            fout.write('</BR>\n')
    
    fout.write('</H1>\n')
    fout.write('</BODY>\n')
    
    fout.write('</HTML>\n')
    
    fout.close()
```

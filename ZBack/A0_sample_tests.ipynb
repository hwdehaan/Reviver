# Words

```python
poem = 'reviver waking—waking is possible—possible is seeing—seeing space is there—there is space in words—works in you can hear—here you can everything stop—stop everything see the violence—violence the sea of filters—filters of and strife—strife and now rest—rest now'

poem = 'one two three four five six seven eight nine ten eleven twelve thirteen fourteen fifteen sixteen'

words = poem.split()
```

# Set Parameters

```python
cgsizerow = 50
cgsizecol = 30

fill_int='_'
fill_ext='.'
len_ext_fill = 0

Nbigrows = 4
Nbigcols = 3



nfontsize=0.75

rmax=int(imgarray.shape[0]/cgsizerow)
cmax=int(imgarray.shape[1]/cgsizecol)


print(str(imgrmax) + " " + str(imgcmax))
print(str(rmax) + " " + str(cmax))
```

# Process Mask

```python
mask=np.zeros((rmax+1,cmax+1),dtype=int)

#print(mask[rmax-1,cmax-1])

rowmask = 0
colmask = 0

for row in range(0,imgrmax):
    if row % cgsizerow == 0:
        colmask = 0
        for col in range(0,imgcmax):            
            if col%cgsizecol==0:               
                if imgarray[row,col,0] > 128:
                    #print(str(rowmask) + " " + str(colmask))
                    mask[rowmask,colmask]=0
                    pix = 0
                else:
                    pix = 1
                    mask[rowmask,colmask]=1

                colmask += 1
                #print(pix,end='')
        rowmask+=1
        #print('')

#rmax=mask.shape[0] - 1
#cmax=mask.shape[1] - 1
```

# Make text and write HTML

```python


fout = open('shape_1.html','w')

fout.write('<HTML>\n')

fout.write('<HEAD>\n')
fout.write('<style> h1 {font-size: ' + str(nfontsize) + 'em; font-family: courier; font-style: regular; font-weight: normal;)</style>\n')
fout.write('</HEAD>\n')

fout.write('<BODY>\n')
fout.write('<H1>\n')



curr_shape=0

full_text=[]
i_inword = np.zeros(len(words),dtype=int)

iword = 0
for kk in range(0,Nbigrows):
    for i in range(0,rmax):
        this_line=''
        for k in range(0,Nbigcols):           
            for j in range(0,len_ext_fill):
                this_line = this_line + fill_ext
            for j in range(0,cmax):
                itile = kk*Nbigcols+k
                if mask[i,j] == 0:
                    this_line = this_line + fill_int
                else:
                    this_line = this_line + words[itile][i_inword[itile]]
                    i_inword[itile] += 1
                    if i_inword[itile] >= len(words[itile]):
                        i_inword[itile]=0
            for j in range(0,len_ext_fill):
                this_line = this_line + fill_ext
        print(this_line)
        full_text.append(this_line)

        
        for ijk in range(0,len(this_line)):
            fout.write(this_line[ijk])
        fout.write('</BR>\n')

fout.write('</H1>\n')
fout.write('</BODY>\n')

fout.write('</HTML>\n')

fout.close()
```
